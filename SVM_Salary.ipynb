#Feature Scaling

```python
x_train=traindf.drop(['Salary'],axis=1)
y_train=traindf['Salary']
x_test=testdf.drop(['Salary'],axis=1)
y_test=testdf['Salary']
```

```python
sc=StandardScaler()
```

```python
x_train=pd.DataFrame(sc.fit_transform(x_train),columns=x_train.columns)
x_test=pd.DataFrame(sc.fit_transform(x_test),columns=x_test.columns)
```

#SVM model

```python
#linear kernel
model=SVC(kernel='linear',C=15)
```

```python
model.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)
```

```python
accuracy_score(y_test,ypred)
```

```python
print(classification_report(y_test,ypred))
```

```python
#polynominal kernel
model1=SVC(kernel='poly',C=15)
```

```python
model1.fit(x_train,y_train)
```

```python
ypred1=model1.predict(x_test)
```

```python
accuracy_score(y_test,ypred1)
```

```python
print(classification_report(y_test,ypred1))
```

```python
#sigmoid kernel
model2=SVC(kernel='sigmoid',C=15)
```

```python
model2.fit(x_train,y_train)
```

```python
ypred2=model2.predict(x_test)
```

```python
accuracy_score(y_test,ypred2)
```

```python
print(classification_report(y_test,ypred2))
```

```python
#rbf kernel
model3=SVC(kernel='rbf',C=15)
model3.fit(x_train,y_train)
ypred3=model3.predict(x_test)
```

```python
accuracy_score(y_test,ypred3)
```

```python
print(classification_report(y_test,ypred3))
```
